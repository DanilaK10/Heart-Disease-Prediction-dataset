# Importing the data

Down below it imports all the libaries needed for the tasks and it also retrieves the dataset.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')
import plotly.graph_objs as go
import plotly.offline as py
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/DanilaK10/Heart-Disease-Prediction-dataset/main/Heart%20Disease%20Dataset.csv")
df
```

# 3.2 Renaming Columns

Renames the columns in the dataset - it is required later on with some of the tasks.

```python
#renaming columns - Kaggle - 3.2
df.columns = ["Age", "Sex", "Chest_pain_type", "Resting_bp", "Cholesterol", "Fasting_bs", "Resting_ecg", "Max_heart_rate", "Exercise_induced_angina", "ST_depression", "ST_slope", "Num_major_vessels", "Thallium_test", "Condition"]
df
```

# Splitting the data

x is for all the values from the dataset except whether the patient has Heart disease or not. This data will then be used to train the model with the y which contains the corresponding values for whether that particular patient has heart disease or not.

>

test_size represents the amount of data left over from the dataset for x_test and y_test (0.2 means 20%) and random_state randomizes the order of the data from the dataset.


```python
y = df["Condition"] # - y is the condition values from the dataset
x = df.drop("Condition", axis = 1) # - x is the dataset with condition removed
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=23)
x_test
```

# Normalization of test + train data


>

https://stats.stackexchange.com/questions/287425/why-do-you-need-to-scale-data-in-knn

```python
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
normalized_x_train = scalar.fit_transform(x_train)
normalized_x_test = scalar.fit_transform(x_test)
```

# KNeighbors Classification

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier()
KNN.fit(normalized_x_train,y_train)
y_pred_KNN = KNN.predict(normalized_x_test)
```

# GridSearch to find best paramaters for Kneighbor Classifier
n_neighbor = 10
>
p = 1
>
weights = distance
>
algorithm + leaf_size dont seem to change accuracy






```python
from sklearn.model_selection import GridSearchCV
k_range = list(range(9,15))
l_size = list(range(1,5))
paramater_grid = dict(n_neighbors = k_range, weights = ["distance"],algorithm = ["ball_tree"], leaf_size = l_size,p = [1])
grid = GridSearchCV(KNN, paramater_grid, cv=5, scoring = "accuracy")
grid.fit(normalized_x_train,y_train)
pd.DataFrame(grid.cv_results_)[['mean_test_score', 'std_test_score', 'params']]
```

```python
print(grid.best_score_)
print(grid.best_params_)
#print(normalized_train_x)
```
