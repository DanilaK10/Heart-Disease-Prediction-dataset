# Importing the data

Down below it imports all the libaries needed for the tasks and it also retrieves the dataset.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')
import plotly.graph_objs as go
import plotly.offline as py
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/DanilaK10/Heart-Disease-Prediction-dataset/main/Heart%20Disease%20Dataset.csv")
df
```

# 3.2 Renaming Columns

Renames the columns in the dataset - it is required later on with some of the tasks.

```python
#renaming columns - Kaggle - 3.2
df.columns = ["Age", "Sex", "Chest_pain_type", "Resting_bp", "Cholesterol", "Fasting_bs", "Resting_ecg", "Max_heart_rate", "Exercise_induced_angina", "ST_depression", "ST_slope", "Num_major_vessels", "Thallium_test", "Condition"]
df
```

# Splitting the data

x is for all the values from the dataset except whether the patient has Heart disease or not. This data will then be used to train the model with the y which contains the corresponding values for whether that particular patient has heart disease or not.

>

test_size represents the amount of data left over from the dataset for x_test and y_test (0.2 means 20%) and random_state randomizes the order of the data from the dataset.


```python
y = df["Condition"] # - y is the condition values from the dataset
x = df.drop("Condition", axis = 1) # - x is the dataset with condition removed
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.4, random_state=25)
x_test
```

# Machine - Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter = 1000)
lr.fit(x_train, y_train)
y_pred_LR = lr.predict(x_test)
```

accuracy

```python
y_testArray = np.array(y_test.explode())
count = 0
for i in range(len(y_testArray)):
  if str(y_pred_LR[i]) == str(y_testArray[i]):
    count += 1
avg = (count/len(y_pred_LR))*100
print("Number of right predictions : " + str(count))
print("Total number of predictions : " + str(len(y_pred_LR)))
print("Percentage of right predictions : " + str(avg))
```

# LinearSVM
LinearSVC uses LibLinear that uses a random number generator therefore it can be ran multiple times on the same input data and have different output -- you can set max_iter = 1000000 which from testing, it seems to fix this issue and the outputs are now constant and to remove the ConvergenceWarning you can set max_iter = 10000000 (takes a minute to run)



```python
from sklearn import svm
SVC = svm.LinearSVC(max_iter = 10000000)
SVC.fit(x_train,y_train)
y_pred_SVC = SVC.predict(x_test)
```

accuracy

```python
y_testArray = np.array(y_test.explode())
count = 0
for i in range(len(y_testArray)):
  if str(y_pred_SVC[i]) == str(y_testArray[i]):
    count += 1
avg = (count/len(y_pred_SVC))*100
print("Number of right predictions : " + str(count))
print("Total number of predictions : " + str(len(y_pred_SVC)))
print("Percentage of right predictions : " + str(avg))
```

# KNeighbors Classification

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=13)
KNN.fit(x_train,y_train)
y_pred_KNN = KNN.predict(x_test)
#y_pred
```

accuracy
